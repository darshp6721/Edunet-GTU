# 1. Dictionary Methods

```python
student = {
    "name": "Rahul",
    "age": 20,
    "course": "AI"
}

print(student)

# 1. get()
print(student.get("name"))

# 2. keys()
print(student.keys())

# 3. values()
print(student.values())

# 4. items()
print(student.items())

# 5. update()
student.update({"city": "Ahmedabad"})
print(student)
```

# List Methods

```python
fruits = ["Apple", "Banana", "Mango"]

# 1. append()
fruits.append("Orange")
print(fruits)

# 2. insert()
fruits.insert(1, "Grapes")
print(fruits)

# 3. remove()
fruits.remove("Banana")
print(fruits)

# 4. pop()
fruits.pop()
print(fruits)

# 5. sort()
numbers = [5, 2, 9, 1]
numbers.sort()
print(numbers)
```

Tuple Methods

```python
numbers = (10, 20, 30, 20, 40)

# 1. count()
print(numbers.count(20))

# 2. index()
print(numbers.index(30))

# 3. len()
print(len(numbers))

# 4. max()
print(max(numbers))

# 5. min()
print(min(numbers))
```

Set Methods

```python
colors = {"Red", "Blue", "Green"}

# 1. add()
colors.add("Yellow")
print(colors)

# 2. remove()
colors.remove("Blue")
print(colors)

# 3. copy()
new_colors = colors.copy()
print(new_colors)

# 4. union()
set2 = {"Black", "White"}
print(colors.union(set2))

# 5. clear()
temp = {"A", "B"}
temp.clear()
print(temp)
```

## If Statement

```python
age = 20

if age >= 18:
    print("Eligible to vote")
```

## If-Else Statement

```python
number = 7

if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

## If-Elif-Else Statement

```python
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
```

## Nested-If-Else Statement

```python
age = 22
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to vote")
    else:
        print("Not a citizen")
else:
    print("Too young")
```

## Break Statement

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

## Continue Statement

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

## Pass Statement

```python
for i in range(3):
    if i == 1:
        pass
    print(i)
```
