## Creating a Multidimensional Array

```python
arr = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])

print(arr)
```

## Accessing Values Using Index Numbers

```python
print(arr[0][0])
```

```python
print(arr[1][2])
```

```python
print(arr[2][1])
```

## Performing Mathematical Operations:

##Addition:

```python
print(arr + 10)
```

## Multiplication:

```python
print(arr * 2)
```

## Division:

```python
print(arr / 5)
```

## Square:

```python
print(arr ** 2)
```

## 15 NumPy Methods

```python
#1. shape
print(arr.shape)

#2. ndim
print(arr.ndim)

#3. size
print(arr.size)

#4. dtype
print(arr.dtype)

#5. reshape
print(arr.reshape(1,9))

#6. flatten
print(arr.flatten())

#7. transpose
print(arr.T)

#8. sum
print(np.sum(arr))

#9. mean
print(np.mean(arr))

#10. max
print(np.max(arr))

#11. min
print(np.min(arr))

#12. sort
x = np.array([8,2,9,4])
print(np.sort(x))

#13. sqrt
print(np.sqrt(arr))

#14. unique
x = np.array([1,2,2,3,4,4,5])
print(np.unique(x))

#15. concatenate
a = np.array([1,2,3])
b = np.array([4,5,6])
print(np.concatenate((a,b)))
```
