## First Analysis

# Dataset Overview

This dataset represents sustainability-related data collected from a smart campus. It includes information about energy consumption, renewable energy generation, water usage, waste production, weather conditions, occupancy levels, and energy savings.

The dataset can be used to analyze how environmental conditions and human activities affect resource utilization and sustainability.

## Feature (Column) Analysis

v# Feature Analysis

### Building_ID
Represents the building where the data was collected.

### Water_Usage_Liters
Shows the amount of water consumed by the building.

### Waste_Produced_Kg
Represents the total waste generated.

### Recycled_Waste_Kg
Shows how much waste was recycled.

### Occupancy_Level
Number of people occupying the building.

### Temperature_C
Ambient temperature in degrees Celsius.

### Humidity_%
Percentage of humidity.

### Cloud_Cover_%
Cloud coverage affecting sunlight availability.

### Wind_Speed_kmph
Wind speed in kilometers per hour.

### Dew_Point_C
Temperature at which water vapor condenses.

### Precipitation_mm
Rainfall measured in millimeters.

### Beam_Irradiance_Wm2
Direct solar radiation received.

### Diffuse_Irradiance_Wm2
Scattered solar radiation.

### Reflected_Irradiance_Wm2
Solar radiation reflected from surrounding surfaces.

### Is_Peak_Hour
Indicates whether the data belongs to peak electricity usage hours.

### Day_Type
Represents whether it is a weekday or weekend.

### Solar_Generation_kWh
Electricity generated using solar panels.

### Energy_Consumption_kWh
Total electricity consumed.

### Total_Energy_Saved_kWh
Amount of energy saved through sustainable practices.

# Checking Null Values

```python
df.isnull().sum()
```

## Handling Null Values

```python
if df.isnull().sum().sum() == 0:
    print("No missing values found in the dataset.")
else:
    df.fillna(df.mean(numeric_only=True), inplace=True)
```

## About Missing Values

# Handling Missing Values

The dataset was checked for missing (null) values using the `isnull()` function.

No missing values were found in any column.

Therefore, no data cleaning or replacement of values was required.

## Relationship Analysis:

# Relationship Between Features

- Higher occupancy generally leads to increased water usage and higher energy consumption.
- Buildings producing more solar energy can reduce overall energy demand from conventional electricity sources.
- Waste produced and recycled waste are positively related because more waste provides more opportunities for recycling.
- Weather conditions such as cloud cover directly influence solar energy generation.
- Beam irradiance has a positive effect on solar power generation.
- Energy saved depends on both energy consumption and renewable energy production.
- Peak hours usually result in higher electricity consumption.

## General Observations:

# General Observations

- The dataset contains sustainability information from multiple campus buildings.
- Renewable energy generation contributes to reducing electricity demand.
- Environmental factors influence solar power production.
- Efficient waste recycling supports sustainable campus management.
- Occupancy level affects both water usage and energy consumption.
- Smart campuses can reduce energy usage by combining renewable energy with efficient resource management.

## Visualizations:

Energy Consumption Histogram

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df["Energy_Consumption_kWh"], bins=20)
plt.title("Energy Consumption Distribution")
plt.xlabel("Energy Consumption (kWh)")
plt.ylabel("Frequency")
plt.show()
```

Solar Generation vs Energy Consumption:

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Solar_Generation_kWh"], df["Energy_Consumption_kWh"])
plt.title("Solar Generation vs Energy Consumption")
plt.xlabel("Solar Generation (kWh)")
plt.ylabel("Energy Consumption (kWh)")
plt.show()
```
